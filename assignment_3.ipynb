**1**. **Read mobile dataset using pandas, and preprocess it (fill null values)**.

```
#filling null values with mean
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/drive/My Drive/mobile_cleaned-1549119762886 (2).csv')
weight_avg=round(df["weight"].mean(),2)
stand_by_time_avg=round(df["stand_by_time"].mean(),2)
no_of_reviews_in_gsmarena_in_week_avg=round(df["no_of_reviews_in_gsmarena_in_week"].mean(),2)
phone_height_avg=round(df["phone_height"].mean(),2)
talk_time_avg=round(df["talk_time"].mean(),2)
video_resolution_avg=round(df["video_resolution"].mean(),2)
primary_camera_resolution_avg=round(df["primary_camera_resolution"].mean(),2)

#  Replaceing the 'NaN' to column mean values

df["weight"].fillna(weight_avg, inplace=True)
df["stand_by_time"].fillna(stand_by_time_avg, inplace=True)
df["no_of_reviews_in_gsmarena_in_week"].fillna(no_of_reviews_in_gsmarena_in_week_avg, inplace=True)
df["phone_height"].fillna(phone_height_avg, inplace=True)
df["talk_time"].fillna(talk_time_avg, inplace=True)
df["video_resolution"].fillna(video_resolution_avg , inplace=True)
df["primary_camera_resolution"].fillna(primary_camera_resolution_avg, inplace=True)
df[100:]
```

**3**.**Compute outer product of 2 vectors**.

```
import numpy as np
matrix1 = [[1, 0], [0, 1]]
matrix2 = [[1, 2], [3, 4]]
print("original matrix:")
print(matrix1)
print(matrix2)
result = np.outer(matrix1, matrix2)
print("Outer product of the said two vectors:")
print(result)
```

**4**. **Write a NumPy program to compute the condition number of a given matrix**.

```
import numpy as np
from numpy import linalg as na
matrix= np.array([[1,2],[3,4]])
print("Original matrix:")
print(matrix)
print("The condition number of the said matrix:")
print(na.cond(matrix))
```

**5**. **write a NumPy program to compute the sum of the diagonal element of a given array(matrix)**.

```
import numpy as np
n_array=np.array([[4,2,1],
                  [3,4,2],
                  [1,5,4]])
print("matrix is:")
print(n_array)
sum=np.trace(n_array)
print("sum of the diagonal elements :",sum)
```
